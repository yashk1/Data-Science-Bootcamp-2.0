# Z test

```python
stats.norm.ppf(0.025)
```

```python
from numpy.random import randn
population  = randn(100)
```

```python
sns.displot(population, kde = True)
```

```python
print(f'Mean: {np.mean(population)}, Std Dev: {np.std(population)}')
```

```python
null_mean = 0.031781337219746096
```

alpha = 0.05
c.i = 95%

p<=0.05 #reject the null hypothesis

p> 0.05 #accept teh null hypothesis

```python
from statsmodels.stats.weightstats import ztest

alpha = 0.05
zscore, p_value = ztest(population, value = null_mean, alternative = 'larger')
```

```python
p_value
```

```python
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print('accept the alternate hypothsis')
```
