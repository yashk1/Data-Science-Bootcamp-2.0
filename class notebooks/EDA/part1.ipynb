```python
#importing libraries
import pandas as pd 
```

```python
df_copy['reviews'] = df_copy['reviews'].astype('int')
```

```python
df_copy["size"].dtype
```

```python
df_copy["size"].unique()
```

```python
df_copy["size"] = df_copy["size"].str.replace("M","000")
df_copy["size"] = df_copy["size"].str.replace("k","")
```

```python
df_copy["size"].unique()
```

```python
import numpy as np
df_copy["size"] = df_copy["size"].str.replace('Varies with device',str(np.nan))
```

```python
df_copy["size"].unique()
```

```python
df_copy["size"] = df_copy["size"].astype("float")
```

```python
df_copy["size"].head()
```

```python
for i in df_copy['size']:
    if i < 10:
        df_copy['size'] = df_copy['size'].replace(i, i*1000)

df_copy['size']
```

```python
df_copy.head()
```

```python
df_copy['installs']=df_copy['installs'].str.replace("+", "")
```

```python
df_copy.price.unique()
```

```python
df_copy.price.dtype
```

```python
char_to_removes=["+",",","$"]
cols_to_clean=["installs","price"]

for item in char_to_removes:
    for col in cols_to_clean:
        df_copy[col]=df_copy[col].str.replace(item,"")
```

```python
df_copy.head()
```

```python
df_copy.info()
```

```python
df_copy['installs'] = df_copy['installs'].astype('int')
df_copy['price'] = df_copy['price'].astype('float')
```

```python
df_copy.head()
```

```python
df_copy.last_updated = pd.to_datetime(df_copy['last_updated'])
df_copy.head()
```

```python
df_copy['year'] = df_copy.last_updated.dt.year
df_copy['month'] = df_copy.last_updated.dt.month
df_copy['day'] = df_copy.last_updated.dt.day

```

```python
df_copy.head()
```

```python
df_copy["android_ver"]=df_copy["android_ver"].str.replace("and up","")
df_copy["android_ver"]=df_copy["android_ver"].str.replace("Varies with device","NaN")
```

```python
df_copy.to_csv("clean_google_data", index= False)
```
